```python
import time
import pandas as pd
import numpy as np

global Filter 
CITY_DATA = { 'chicago': 'chicago.csv',
              'new york city': 'new_york_city.csv',
              'washington': 'washington.csv' }
```

```python
def get_filters():
    """
    Asks user to specify a city, month, and day to analyze.

    Returns:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    """
    print('Hello! Let\'s explore some US bikeshare data!')
    # get user input for city (chicago, new york city, washington). HINT: Use a while loop to handle invalid inputs
    city = byCity()       
    
    # get user input for month (all, january, february, ... , june)
    #listOfGlobals = globals()
    #listOfGlobals['Filter'] = input('Would you like to filter the data by month, day, both or not at all? Type none for no time filter \n>').lower()
    Filter = input(print('Would you like to filter the data by month, day, both or not at all? Type none for no time filter \n>')).lower()
    
    if Filter == "month":
        month = byMonth()
    if Filter == "day" :   # get user input for day of week (all, monday, tuesday, ... sunday)
        print('Please enter month to filter the data by day')
        month = byMonth()
        day = byDay()
    if Filter == "both":
        month = byMonth()
        day = byDay()
    else:                          
        month = day = "all" 
        
    print('-'*40)
    return city, month, day
```

```python
def byCity():   
    '''Ask User to enter a city
       
       Arguments : Nothing
       
       Returns : A city name
    
    '''
    city = ''
    while city not in CITY_DATA.keys():
        # Taking input from user and converting into the lower to standardize them.
        city = input(print('Would you like to see data for Chicago, New York city, or Washington? \n>')).lower()
        
        if city not in CITY_DATA.keys():
            print("\nPlease check your input, it doesn\'t appear in any of the accepted input formats.")
            print("\nRestarting...")
       
    
    return city
```

```python
def byMonth():      #function to get input for month
    '''Ask User to enter a month
       
       Arguments : Nothing
       
       Returns : A month name
    
    '''
    month = ''
    #Creating a dictionary to store all the months including the 'all' option
    MONTH_DATA = {'january': 1, 'february': 2, 'march': 3, 'april': 4, 'may': 5, 'june': 6, 'all': 7}
    
    while month not in MONTH_DATA.keys():
        # Taking input from user and converting into the lower to standardize them.
        month = input(print('Which month? January, Febuary, March, April, May or June \n>')).lower()
    
        if month not in MONTH_DATA.keys():
            print('\n Please try again. Input is not in accepted format\n')
            print('Restarting...')
    
    return month
```

```python
def byDay():      
    '''Ask User to enter a day in Integer format
       
       Arguments : Nothing
       
       Returns : A Day Number
    
    '''
    day = ''
    DAY_DATA = {0 : 'sunday', 1 : 'monday', 2 : 'tuesday', 3 : 'wednesday', 4: 'thursday', 5 : 'friday', 6 : 'saturday', 7 : 'all'}
    
    while day not in DAY_DATA.keys(): 
        day = int(input(print('Which day? Please type your response as an integer (e.g 0 = Sunday) \n>')))
    
        if day not in DAY_DATA.keys(): 
            print("\nInvalid input. Please try again in the accepted input formats.")
            print("\nRestarting...")
        
    return day
```

```python
def load_data(city, month, day):
    """
    Loads data for the specified city and filters by month and day if applicable.

    Args:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    Returns:
        df - Pandas DataFrame containing city data filtered by month and day
    """
    # Load CSV file for specifed city
    df = pd.read_csv(CITY_DATA[city])
    
    # extract month from Start Time to create new columns
    df['Start Time'] = pd.to_datetime(df['Start Time'])
    df['month'] = df['Start Time'].dt.month
    
    # extract day of week from Start Time to create new columns
    df['day_of_week'] = df['Start Time'].dt.strftime("%A")
    
    # Filter dataframe by month
    if month != 'all' :
        months = ['january', 'february', 'march', 'april', 'may', 'june']
        month = months.index(month)+1 
        df = df[df['month'] == month]

    # Filter dataframe by day of week
    if day != 'all' :  
        days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
        df = df[df['day_of_week'] == days[day]]

    return df
```

```python
def time_stats(df):
    """Displays statistics on the most frequent times of travel."""

    print('\nCalculating The Most Frequent Times of Travel...\n')
    time.sleep(10)
    start_time = time.time()
    
    
    # display the most common month
    months = ['january', 'february', 'march', 'april', 'may', 'june']
    common_month = df['month'].mode().index[0]
    count = df['month'].mode().max()
    print("Common Month : {} Count : {} Filter : {} ".format(months[common_month].title(), count, Filter))
    
    # display the most common day of week
    common_day = df['day_of_week'].value_counts().index[0]
    count = df['day_of_week'].value_counts()[0]
    print("Common Day of Week : {} Count : {} Filter : {} ".format(common_day, count, Filter))
    
    # display the most common start hour
    df['hour'] = df['Start Time'].dt.hour
    common_hour = df['hour'].mode().index[0]
    count = df['hour'].mode().max()
    print("Common Hour : {} Count : {} Filter : {} ".format(common_hour, count, Filter))
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```

```python
def station_stats(df):
    """Displays statistics on the most popular stations and trip."""

    print('\nCalculating The Most Popular Stations and Trip...\n')
    time.sleep(10)
    start_time = time.time()

    # display most commonly used start station
    common_start_station = df['Start Station'].value_counts().index[0]
    count = df['Start Station'].value_counts()[0]
    
    print("Common Start Station : {} Count : {} Filter : {} ".format(common_start_station, count, Filter))
    
    # display most commonly used end station
    common_end_station = df['End Station'].value_counts().index[0]
    count = df['End Station'].value_counts().max()
    print("Common end Station : {} Count : {} Filter : {} ".format(common_end_station, count, Filter))

    # display most frequent combination of start station and end station trip
    common_start_end_station = df[['Start Station', 'End Station']].mode().index[0]
    count = df[['Start Station', 'End Station']].mode().max()
    print("most frequent combination of start station and end station trip : {} Count : {} Filter : {} ".format(common_start_end_station, count, Filter))
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```

```python
def trip_duration_stats(df):
    """Displays statistics on the total and average trip duration."""

    print('\nCalculating Trip Duration...\n')
    time.sleep(10)
    start_time = time.time()

    # display total travel time
    total_travel_time = df['Trip Duration'].sum()
    print("Total travel time : {}".format(total_travel_time))
    # display mean travel time
    total_travel_time = df['Trip Duration'].mean()
    print("Total travel time : {}".format(total_travel_time))

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```

```python
def user_stats(df):
    """Displays statistics on bikeshare users."""

    print('\nCalculating User Stats...\n')
    time.sleep(10)
    start_time = time.time()

    # Display counts of user types
    if 'User Type' in df.columns:
        user_stats_type(df)
        
    # Display counts of gender
    if 'Gender' in df.columns:
        user_stat_gender(df)

    # Display earliest, most recent, and most common year of birth
    if 'Birth Year' in df.columns:
        user_stat_birth(df)

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```

```python

def user_stats_type(df): 
    '''
    Display counts of user types
    
    Arguments : Dataframe
    
    Returns : Nothing
    
    '''
    user_types = df['User Type'].value_counts()
    for index,user_type in enumerate(user_types):  
        print("{} : {} Filter : {}".format(user_types.index[index], user_type, Filter)) 
```

```python
def user_stat_gender(df):     
    '''
    Display counts of gender
    
    Arguments : Dataframe
    
    Returns : Nothing
    
    '''
    gender_counts = df['Gender'].value_counts()
    for index, gender_count in enumerate(gender_counts):
        print("{} : {} Filter : {} ".format(gender_counts.index[index], gender_count, Filter))
```

```python
def user_stat_birth(df):
    '''
    Display earliest, most recent, and most common year of birth
    
    Arguments : Dataframe
    
    Returns : Nothing
    
    '''
    # earliest birth year
    most_earliest_year = df['Birth Year'].min()
    print("The most earliest birth year : ", most_earliest_year)
    
    # most recent birth year
    most_recent_year = df['Birth Year'].max()
    print("The most recent birth year : ", most_recent_year)
    
    # most common birth year
    most_common_year = df['Birth Year'].value_counts().idxmax()
    print("The most common birth year : ", most_common_year)
    
```
